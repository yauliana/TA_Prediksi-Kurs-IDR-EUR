```python
import csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statistics
import datetime
import joblib
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import make_scorer
from sklearn.metrics import r2_score
import math
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
```

```python
df = pd.read_csv("/content/drive/ My Drive/Colab_TA/Dataset_idr.csv", usecols=["Date", "Units per EUR"])
df['Date']= pd.to_datetime(df.Date)
df
```

```python
x=np.array(range(273))
x=x.reshape(-1,1)
df['day'] = x+1
```

```python
x=df.drop(['Date','Units per EUR'],axis = 1)
y=df['Units per EUR']
features = x.columns
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
```

```python
scorer = make_scorer(mean_absolute_error, greater_is_better= False)

parameters_kernel = {'C': [0.001,0.1, 1],
                     'gamma' :[1],
                     'epsilon': [0.01,0.1,1],
                     'coef0':[1,2,3],
                     "degree":[2],
                     'kernel' :['poly']}

svr = GridSearchCV(SVR(), parameters_kernel, cv=2, scoring= scorer)
svr.fit(x_train, y_train)
```

```python
means = svr.cv_results_['mean_test_score']
print("best Params = ", svr.best_params_)
```

```python
model_svr = SVR(kernel='poly', C=0.1 ,coef0=2, degree=2, epsilon=1, gamma=1)
model_svr.fit(x_train,y_train)
```

```python
y_pred_test= model_svr.predict(x_test)
y_pred_train = model_svr.predict(x_train)
```

```python
print('R^2 Train : {0:.5f}'.format(r2_score(y_train, y_pred_train)))
print('R^2 Test: {0:.5f}'.format(r2_score(y_test, y_pred_test)))
```

```python
filename = 'model_SVR.sav'
joblib.dump(model_svr, filename)
```
